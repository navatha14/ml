```python
# Subset rows where an external array, bar, is positive

bar = np.array([2,-1,9,-10,1,4,-1,-10,37,8])
subset_bar = df[bar > 0]
subset_bar
```

```python
# 8. Subset rows where foo is TRUE or bar is negative

bar = np.array([2,-1,9,-10,1,4,-1,-10,37,-8])
foo = np.array([False,True,False,True,False,False,True,False,False,False])

subset_foobar = df[ (foo)| bar < 0]
subset_foobar
```

```python
 # Subset the rows where foo is not TRUE and bar is not negative

bar = np.array([2,-1,9,-10,1,4,-1,-10,37,-8])
foo = np.array([False,True,False,True,False,False,True,False,False,False])

subset_nefoobar = df[ ~(foo) & ~(bar < 0)]
subset_nefoobar
```
